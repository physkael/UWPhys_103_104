```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
x = np.arange(-1.2, 1.2, 0.01)
x_0 = 0.65
plt.plot(x, np.sin(2*np.pi*(x - x_0)))
plt.plot(x, np.sin(2*np.pi*(x + x_0)))
plt.grid(1)
```

```python
fig, ax = plt.subplots()
w1, = ax.plot([], [])
w2, = ax.plot([], [])
phase_annotation = ax.text(0.5, 0.8, '', horizontalalignment='center', fontsize=14, fontweight=300)
ax.set_xlim((-1.25, 1.25))
ax.set_ylim((-2, 2))
ax.grid(True)

def wave_interference(x0):
    x = np.arange(-1.25, 1.25, 0.01)
    w1.set_data(x, np.sin(2*np.pi*(x - x0)))
    w2.set_data(x, np.sin(2*np.pi*(x + x0)))
    phase_annotation.set_text(rf'$\phi = {x0:.2f}\;\lambda$')
    return w1, w2

# Only save last 100 frames, but run forever
ani = animation.FuncAnimation(fig, func=wave_interference, frames=np.linspace(-1, 1, 101), interval=100)
ani.save('waveint.gif')
#plt.show()
```
